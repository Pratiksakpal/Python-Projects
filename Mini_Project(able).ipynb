```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv('/content/owid-covid-data.csv')
data.head()
```

```python
# Basic Summary of the Data

data.describe()
```

```python
# Filter the DataFrame to include only rows where total_deaths is greater than 1,000,000
high_death_countries = data[data['total_deaths'] > 1000000]
```

```python
# Count the number of rows in the filtered DataFrame
total_countries_with_high_deaths = high_death_countries.shape[0]

print("Total countries with total deaths > 1,000,000:", total_countries_with_high_deaths)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtering the dataset based on the condition total_deaths > 1000000
filtered_data = data[data['total_deaths'] > 1000000]

# Grouping the data by countries where total_deaths is greater than 1000000
grouped_data = filtered_data.groupby('location')

# Count the number of countries in each group (location)
country_counts = grouped_data.size()

# Create a bar chart to visualize the data
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Location')
plt.ylabel('Total deaths in 1000s')
plt.title('Countries with Total Deaths > 1,000,000')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```
